```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load 
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
print(school_data)
school_data.head()
student_data = pd.read_csv(student_data_to_load)
print(student_data)
student_data.head()
# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Create a list of unique school name
just_school_names = school_data['school_name'].unique()
#Total number of unique schools
count_of_schools = len(just_school_names)
#Total number of students in the district
count_of_students_in_district = school_data['size'].sum()
#Total Budget
Total_budget = school_data['budget'].sum()
Average_math_score = student_data['math_score'].mean()
#Calculate the average reading score
Average_reading_score = student_data['reading_score'].mean()
#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
overall_average_score = (Average_math_score + Average_reading_score) / 2
#Calculate the percentage of students with a passing math score (70 or greater)
Passing_math_score_70 = student_data[(student_data['math_score'] >= 70)]['student_name'].count()/count_of_students_in_district*100
#Calculate the percentage of students with a passing reading score (70 or greater)
Passing_reading_score_70 = student_data[(student_data['reading_score'] >= 70)]['student_name'].count()/count_of_students_in_district*100
#Create a dataframe to hold the above results
district_summary = pd.DataFrame({"Total Schools":[count_of_schools],
                                "Total Students":[count_of_students_in_district],
                                "Total Budget":[Total_budget],
                                "Average Math Score":[Average_math_score],
                                "Average Reading Score":[Average_reading_score],
                                "%Passing Math":[Passing_math_score_70],
                                "%Passing Reading":[Passing_reading_score_70],
                                 "%Overall Passing Rate":[overall_average_score]})
#Cleaner format
district_summary.style.format({"Total Budget": "${:,.2f}", 
                       "Average Math Score": "{:.1f}", 
                       "Average Reading Score": "{:.1f}", 
                       "% Passing Math": "{:.1%}", 
                       "% Passing Reading": "{:.1%}", 
                       "Overall Passing Rate": "{:.1%}"})
```

# School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
#Schools grouped together
School_name_grouping = school_data_complete.set_index('school_name').groupby(['school_name'])
#school types
school_types = school_data.set_index('school_name')['type']
# total students by school
students_per_school = School_name_grouping['Student ID'].count()
# school budget
each_school_budget = school_data.set_index('school_name')['budget']
#student wise budget
individual_student_budget = school_data.set_index('school_name')['budget']/school_data.set_index('school_name')['size']
#avg scores by school
average_math_score = School_name_grouping['math_score'].mean()
average_read_score = School_name_grouping['reading_score'].mean()
# % passing scores
percent_passing_math = school_data_complete[school_data_complete['math_score'] >= 70].groupby('school_name')['Student ID'].count()/students_per_school
percent_passing_read = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('school_name')['Student ID'].count()/students_per_school
overall_passing_rate = (percent_passing_math + percent_passing_read) / 2
school_summary = pd.DataFrame({
    "School Name": School_name_grouping,
    "School Type": school_types,
    "Total Students": students_per_school,
    "Total School Budget": each_school_budget,
    "Per Student Budget": individual_student_budget,
    "Average Math Score": average_math_score,
    "Average Reading Score": average_read_score,
    "% Passing Math": percent_passing_math,
    "% Passing Reading": percent_passing_read,
    "Overall Passing Rate": overall_passing_rate})

#Data munging
school_summary = school_summary[['School Type', 
                          'Total Students', 
                          'Total School Budget', 
                          'Per Student Budget', 
                          'Average Math Score', 
                          'Average Reading Score',
                          '% Passing Math',
                          '% Passing Reading',
                          'Overall Passing Rate']]


#Cleaner formatting

school_summary.style.format({'Total Students': '{:,}', 
                          "Total School Budget": "${:,}", 
                          "Per Student Budget": "${:.0f}",
                          'Average Math Score': "{:.1f}", 
                          'Average Reading Score': "{:.1f}", 
                          "% Passing Math": "{: .1%}", 
                          "% Passing Reading": "{: .1%}", 
                          "Overall Passing Rate": "{: .1%}"})
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Create spending bins
bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
school_data_complete['bins'] = pd.cut(school_data_complete['budget']/school_data_complete['size'], bins, labels = group_names)

#group by spending
group_by_spending = school_data_complete.groupby('bins')

#let's do the calculations
average_math_score =group_by_spending['math_score'].mean()
average_read_score = group_by_spending['reading_score'].mean()
perc_passing_math = school_data_complete[school_data_complete['math_score'] >= 70].groupby('bins')['Student ID'].count()/group_by_spending['Student ID'].count()
perc_passing_read = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('bins')['Student ID'].count()/group_by_spending['Student ID'].count()
overall_passing_rate = school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)].groupby('bins')['Student ID'].count()/group_by_spending['Student ID'].count()*100

            
# DataFrame          
scores_by_school_spending = pd.DataFrame({
    "Average Math Score": average_math_score,
    "Average Reading Score": average_read_score,
    '% Passing Math': perc_passing_math,
    '% Passing Reading': perc_passing_read,
    "%Overall Passing Rate": overall_passing_rate})
            
#reorder columns
scores_by_school_spending = scores_by_school_spending[[
    "Average Math Score",
    "Average Reading Score",
    '% Passing Math',
    '% Passing Reading',
    "%Overall Passing Rate"
]]

scores_by_school_spending.index.name = "Spending Ranges Per Student"
scores_by_school_spending = scores_by_school_spending.reindex(group_names)

#formating
scores_by_school_spending.style.format({'Average Math Score': '{:.1f}', 
                              'Average Reading Score': '{:.1f}', 
                              '% Passing Math': '{:.1%}', 
                              '% Passing Reading':'{:.1%}', 
                              'Overall Passing Rate': '{:.1%}'})
```
